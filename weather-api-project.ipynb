### Location Data

```python
location_url = f"http://api.weatherapi.com/v1/current.json?key={my_api_key}&q={city}"
location_response = requests.get(location_url)

location_response.status_code
```

```python
weather_data = location_response.json()
current = weather_data['current']
location = weather_data['location']
temp_f = current['temp_f']
location_summary = (
    f"Weather for {location['name']}, {location['region']}: "
    f"Current temperature: {current['temp_f']}째F / {current['temp_c']}째C."
)

location_summary
```

### Forecast Data

```python
forecast_url = f"{baseurl}forecast.json?key={my_api_key}&q={city}&days=3"
forecast_response = requests.get(forecast_url)

forecast_response.status_code
```

```python
forecast_data = forecast_response.json()
forecast_data
```

```python
forecast_data = forecast_response.json()
forecast = forecast_data['forecast']['forecastday']
forecast
```

```python
forecast_3_days_summary = []
for eachDay in forecast:
    forecast_3_days_summary.append(f"{eachDay['date']}: High {eachDay['day']['maxtemp_f']}F, Low {eachDay['day']['mintemp_f']}F - {eachDay['day']['condition']['text']}")
```

```python
forecast_3_days_summary
```

```python
# Use the datetime library to diplay date in a more appropriate way:
forecast_3_days_summary = []
for eachDay in forecast:
    date_obj = datetime.strptime(eachDay['date'], '%Y-%m-%d')
    formatted_date = date_obj.strftime('%A, %B %#d')
    forecast_3_days_summary.append(f"{formatted_date}:High {eachDay['day']['maxtemp_f']}F / Low {eachDay['day']['mintemp_f']}F - {eachDay['day']['condition']['text']}")
forecast_3_days_summary       
                                       
```
